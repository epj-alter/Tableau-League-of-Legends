```python
import pandas as pd

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 100)
```

```python
worlds = pd.read_csv("../raw-data/Worlds_2019.csv")
```

```python
worlds.head(15)
```

```python
worlds = worlds.drop(columns=["game", "date", "split", "url", "invisiblewardclearrate", "visiblewardclearrate", "patchno"])
```

```python
indexes = worlds.loc[worlds["position"] == "Team"].index
worlds = worlds.drop(indexes)
worlds.head(10)
```

```python
worlds = worlds.replace("\,", ".", regex=True)
worlds.head(5)
```

```python
worlds.to_csv("../data/worlds_2019.csv", index=False)
```
